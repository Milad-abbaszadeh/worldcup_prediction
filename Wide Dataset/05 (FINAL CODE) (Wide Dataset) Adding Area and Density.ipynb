```python
surface1 = []
surface2 = []

for i in range(len(data)):
    try:
        if data['Team1'][i] == 'England':
            surface1.append(130395)
        elif data['Team1'][i] == 'Wales':
            surface1.append(20735)
        elif data['Team1'][i] == 'Scotland':
            surface1.append(80077)
        elif data['Team1'][i] == 'Northern Ireland':
            surface1.append(14130)
        elif data['Team1'][i] == 'Serbia and Montenegro':
            idx1 = area.index[area['Country'] == 'Serbia']
            idx2 = area.index[area['Country'] == 'Montenegro']
            p1 = area['Area'][idx1[0]]
            p2 = area['Area'][idx2[0]]
            surface1.append(p1+p2)
        else:    
            idx1 = area.index[area['Country'] == data['Team1'][i]]
            p1 = area['Area'][idx1[0]]
            surface1.append(p1)
    except:
        print(data['Team1'][i])
        
    try:
        if data['Team2'][i] == 'England':
            surface2.append(130395)
        elif data['Team2'][i] == 'Wales':
            surface2.append(20735)
        elif data['Team2'][i] == 'Scotland':
            surface2.append(80077)
        elif data['Team2'][i] == 'Northern Ireland':
            surface2.append(14130)
        elif data['Team2'][i] == 'Serbia and Montenegro':
            idx1 = area.index[area['Country'] == 'Serbia']
            idx2 = area.index[area['Country'] == 'Montenegro']
            p1 = area['Area'][idx1[0]]
            p2 = area['Area'][idx2[0]]
            surface2.append(p1+p2)
        else:
            idx2 = area.index[area['Country'] == data['Team2'][i]]
            p2 = area['Area'][idx2[0]]
            surface2.append(p2)
    except:
        print(data['Team2'][i])
```
